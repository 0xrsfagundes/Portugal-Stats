```python
import requests

url = "https://dados.gov.pt/s/dadosGovFiles/Crimesregistados20141993portipo.json"

response = requests.get(url)

data = response.json()

print(data)
```

```python
crimes = []
rates = []
years = []

for item in data['d']:
    for year in range(1993,2014):
        crimes.append(item['crime'])
        years.append(year)
        string_helper = "a{}ncrimes".format(year)
        rates.append(item[string_helper])
```

```python
import pandas as pd

data = {'Crime':crimes,'Year':year ,'Rate':rates}

df = pd.DataFrame(data)
df['Rate'] = pd.to_numeric(data['Rate'],errors='coerce')

print(df)
```

```python
grouped_df = df.groupby(['Crime'])['Rate'].sum().reset_index().sort_values(by='Rate', ascending=False)

print(grouped_df)
```

```python
import matplotlib.pyplot as plt

top_10 = grouped_df.nlargest(10,'Rate').reset_index()
top_10.plot(kind='bar',x='Crime',y='Rate')
plt.show()
```
